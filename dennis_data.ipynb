```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dennis.csv')
df = df.drop('UserLanguage', axis=1)
df = df.drop(0)
df.head(12)
```

```python
df.shape
```

```python
stors = ['Storage_low', 'Storage_medium','Storage_high','Storage_insane']
brand = ['Apple','Samsung','OPPO','Xiaomi']
bats = ['Battery_Low','Battery_Medium','Battery_High']
cams = ['Camera_Bad', 'Camera_Medium', 'Camera_Great']
prices = ['Price_low', 'Price_medium', 'Price_high', 'Price_top']
educs = ['High school graduate', 'MBO degree', 'HBO degree',"Bachelor's degree", "Master's degree"]
ages = ['18 - 24', '25 - 34', '45 - 54', '55 - 64', '25 - 44']
dfnew = pd.DataFrame(columns=['id','obs','choiceset','choice', 'Storage', 'Brand', 'Battery','Education','Age', 'Camera', 'Price', 'Sex']+stors+brand+bats+cams+prices+educs+ages)
dfnew.head()
```

```python
persons = []
observ = []
k= 0 
for i in range(df.shape[0]):
    for j in range(10):
        persons.append(i+1)
        persons.append(i+1)
        observ.append(k+1)
        observ.append(k+1)
        
        k+=1
dfnew['id'] = pd.Series(persons)
dfnew['obs'] = pd.Series(observ) 
dfnew.head()
    
```

```python
choices = []
sexes = []
educations = []
age_list = []
for index, row in df.iterrows():
    for q in range(2,12):
        if row['Q%i'%(q)] == 'A':    
            choices.append(1)
            choices.append(0)
        else:
            choices.append(0)
            choices.append(1)
    if row['sex'] == 'Male':
        [sexes.append(1) for j in range(20)]
    elif row['sex'] == 'Female':
        [sexes.append(0) for j in range(20)]
    else:
        [sexes.append(0) for j in range(20)]
    ed = row['educ']
    ag = row['age']
    if ed in educs:
        [educations.append(ed) for j in range(20)]
    else:
        [educations.append('High school graduate') for j in range(20)]
    if ag in ages:
        [age_list.append(ag) for j in range(20)]
    else:
        [age_list.append('18 - 24') for j in range(20)]
    

        
dfnew['choice'] = pd.Series(choices)
dfnew['Sex'] = pd.Series(sexes)
dfnew['Education'] = pd.Series(educations)
dfnew['Age'] = pd.Series(age_list)

dfnew.head()
```

```python
choiceset = []
for j in range(49):
    for i in range(10):
        choiceset.append(i+1)
dfnew['choiceset'] = pd.Series(choiceset)
dfnew.head(20)
```

```python
Storage = []
Brand = []
Battery = []
Camera = []
Price = []
cats = ['Storage', 'Brand', 'Battery', 'Camera', 'Price']
dicter = {}
for cat in cats:
    dicter[cat] = []
for i in range(490):
    for j in range(0,20,2):
        for cat in cats:
            dicter[cat].append(features[cat].iloc[j])
            dicter[cat].append(features[cat].iloc[j+1])
            
# store.append(features['Storage'].iloc[j])
# store.append(features['Storage'].iloc[j+1])
# brand.append(features['Brand'].iloc[j])
# brand.append(features['Brand'].iloc[j+1])
# bat.append(features['Battery'].iloc[j])
# bat.append(features['Battery'].iloc[j+1])
# cam.append(features['Camera'].iloc[j])
# cam.append(features['Camera'].iloc[j+1])
# price.append(features['Price'].iloc[j])
# price.append(features['Price'].iloc[j+1])
for cat in cats:
    dfnew[cat] = pd.Series(dicter[cat])
dfnew.head()   
        
```

```python
for index, row in dfnew.iterrows():
    #storage
    if row['Storage'] == '64GB':
#         dfnew['Storage_low'].loc(index) = 1
        dfnew.loc[dfnew.index[index], 'Storage_low'] = 1

    elif row['Storage'] == '128GB':
#         dfnew['Storage_medium'].loc(index) = 1
        dfnew.loc[dfnew.index[index], 'Storage_medium'] = 1

    elif row['Storage'] == '256GB':
#         dfnew['Storage_high'].loc(index) = 1
        dfnew.loc[dfnew.index[index], 'Storage_high'] = 1

    elif row['Storage'] == '512GB':
#         dfnew['Storage_insane'].loc(index) = 1
        dfnew.loc[dfnew.index[index], 'Storage_insane'] = 1

    #brands
    brand = row['Brand']
    dfnew.loc[dfnew.index[index], brand] = 1
    
    #education
    ed = row['Education']
    dfnew.loc[dfnew.index[index], ed] = 1

    #battery
    s_bat = row['Battery']
    dfnew.loc[dfnew.index[index], 'Battery_%s'%(s_bat)] = 1
    
    #ages
    ag = row['Age']
    dfnew.loc[dfnew.index[index], ag] = 1

    #camera
    s_cat = row['Camera']
#     row['Camera_%s'%(s_cat)].loc(index) = 1
    dfnew.loc[dfnew.index[index], 'Camera_%s'%(s_cat)] = 1
    
    #price
    if row['Price'] == '200-399':
        dfnew.loc[dfnew.index[index], 'Price_low'] = 1

    elif row['Price'] == '400-599':
        dfnew.loc[dfnew.index[index], 'Price_medium'] = 1

    elif row['Price'] == '600-799':
        dfnew.loc[dfnew.index[index], 'Price_high'] = 1

    elif row['Price'] == '800-1099':
        dfnew.loc[dfnew.index[index], 'Price_top'] = 1

        

    
   
    
    

```
